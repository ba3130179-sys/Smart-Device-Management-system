```python
class SmartDevice:
    """Parent class for all smart devices"""

    def __init__(self, device_id, name):
        if not device_id:
            raise ValueError("Device ID cannot be empty")
        self.__device_id = device_id
        self.__power_status = False
        self.name = name

    # Encapsulation:
    @property
    def device_id(self):
        return self.__device_id

    @device_id.setter
    def device_id(self, value):
        if not value:
            raise ValueError("Device ID cannot be empty")
        self.__device_id = value

    # Encapsulation: Getter for power_status, no direct setter
    @property
    def power_status(self):
        return self.__power_status

    def turn_on(self):
        self.__power_status = True
        print(f"{self.name} is now ON.")

    def turn_off(self):
        self.__power_status = False
        print(f"{self.name} is now OFF.")

    def display_info(self):
        status = "ON" if self.__power_status else "OFF"
        print(f"Device ID: {self.__device_id} | Name: {self.name} | Status: {status}")


class SmartThermostat(SmartDevice):
    """Child class - inherits from SmartDevice"""

    def __init__(self, device_id, name, temperature=22):
        super().__init__(device_id, name) # Initialize parent
        self.__temperature = temperature # Private temp reading

    def read_temperature(self):
        if self.power_status:
            print(f"{self.name} temperature: {self.__temperature}°C")
        else:
            print(f"{self.name} is OFF. Turn it on to read temperature.")

    def display_info(self):
        super().display_info()
        print(f"Current Temperature: {self.__temperature}°C")


class SmartLight(SmartDevice):
    """Child class - inherits from SmartDevice"""

    def __init__(self, device_id, name, brightness=50):
        super().__init__(device_id, name)
        self.brightness = brightness # Uses setter for validation

    @property
    def brightness(self):
        return self.__brightness

    @brightness.setter
    def brightness(self, value):
        if 0 <= value <= 100:
            self.__brightness = value
        else:
            raise ValueError("Brightness must be between 0 and 100")

    def adjust_brightness(self, new_brightness):
        if self.power_status:
            self.brightness = new_brightness # Validation happens here
            print(f"{self.name} brightness set to {self.__brightness}%")
        else:
            print(f"{self.name} is OFF. Turn it on first.")

    def display_info(self):
        super().display_info()
        print(f"Brightness: {self.__brightness}%")


class SmartCamera(SmartDevice):
    """Child class - inherits from SmartDevice"""

    def __init__(self, device_id, name):
        super().__init__(device_id, name)
        self.__recording = False

    def start_recording(self):
        if self.power_status:
            self.__recording = True
            print(f"{self.name} started recording.")
        else:
            print(f"{self.name} is OFF. Turn it on to record.")

    def display_info(self):
        super().display_info()
        rec_status = "Recording" if self.__recording else "Not Recording"
        print(f"Camera Status: {rec_status}")


def main():
    # Create devices
    thermostat = SmartThermostat("TH001", "Living Room Thermostat")
    light = SmartLight("LT001", "Bedroom Light")
    camera = SmartCamera("CM001", "Front Door Camera")

    devices = [thermostat, light, camera]

    # Menu-driven interface
    while True:
        print("\n=== Smart Device Management System ===")
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature")
        print("5. Adjust Brightness")
        print("6. Start Recording")
        print("7. Exit")

        choice = input("Select option (1-7): ")

        if choice == "1":
            for device in devices:
                device.display_info()
                print("-" * 30)

        elif choice == "2":
            for device in devices:
                device.turn_on()

        elif choice == "3":
            for device in devices:
                device.turn_off()

        elif choice == "4":
            thermostat.read_temperature()

        elif choice == "5":
            try:
                level = int(input("Enter brightness 0-100: "))
                light.adjust_brightness(level)
            except ValueError as e:
                print(f"Error: {e}")

        elif choice == "6":
            camera.start_recording()

        elif choice == "7":
            print("Exiting Smart Device Management System.")
            break

        else:
            print("Invalid option. Choose 1-7.")


if __name__ == "__main__":
    main()
```
